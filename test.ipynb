# Using BeautifulSoup to taking info anything you want

```python
url = "https://www.musashino-u.ac.jp/"
res = requests.get(url, headers=headers)
#レスポンスのエンコーディングを設定
res.encoding = res.apparent_encoding
# ステータスコードを表示
print (f"ステータスコード: {res.status_code}")
```

```python
#HTMLのソースコードを解析
soup = BeautifulSoup(res.text, 'html.parser')
# soupオブジェクトの型を表示
print(f"soup: {type (soup)}")
```

```python
# ソースコード全体から<a>タグをすべて抽出
soup_a = soup.a
print(f"<a>タグ: {soup_a}")
print(f"<a>タグの型: {type(soup_a)}")
```

```python
# ソースコード全体から<div>タグを抽出
soup_div = soup.div
# print （f"<div>タグ：{soup_div}"）
# print (f"<div>タグの型：{type(soup_div)}")

soup_div_div = soup_div.div
#最初に見つかった<div>タグを表示
print (f"<div>タグの中の<div>タグ：{soup_div_div}")
#<div>タグオブジェクトの型を表示
print (f"<div>タグの中の<div>タグの型：{type(soup_div_div)}")
```

```python
# BeautifulSoupオブジェクトとTagオブジェクトは，"'."でHTMLタグを連結することでより深いタグにアクセスすることができる.
soup_div_div = soup.div.div

# 最初に見つかった<div>タグを表示
print(f"<div>タグの中の<div>タグ：{soup_div_div}")
# <div>タグオブジェクトの型を表示
print(f"<div>タグの中の<div>タグの型：{type(soup_div_div)}")
```

```python
# Tagオブジェクト・stringでタグ内のテキストを取得できる。

# <div>タグ内のテキストを表示
print(f"<div>タグの中の<div>タグのテキスト：{soup_div_div.string}")
```

```python
# 指定したタグが使われている要素のうち、最初に見つかった要素を取得する
h2 = soup.find('h2')
# 最初に見つかった<h2>タグを表示
print (f"{h2}")
# <h2>タグオブジェクトの型を表示
print(f"！変数の型：{type(h2)}です")

```

```python
# 複数のタグを指定して検索することも可能
#例えば、<h2>タグと<h3>タグの両方を検索する場合は以下のようにする。
h2_h3_list = soup.find_all(['h2', 'h3'])
for tag in h2_h3_list:
    print(f"{tag.string}") #各<h2>および<h3>タグ内のテキストを表示
```

```python
list_ = soup.find_all("h2")
for c in list_:
    if c.span is not None:
        print(c.span.string)
    
a_all = soup.find_all("a")
for a in a_all:
    print(a.string)
    
```

```python
# id属性やclass属性など、任意の属性で検索することも可能である。
# 例えば、id属性が'main-content'の要素を検索する場合は以下のようにする。
tag_id_list = soup.find_all(attrs={'id': 'main-content'}, limit=5)
for tag in tag_id_list:
    print(f"{tag}") # id属性が'main-content'のタグを表示
```

```python
href_list = soup.find_all(attrs={"href": "/"})
for tag in href_list:
    print(tag)

list_ = soup.find_all(class_="p-notices")
for c in list_:
    print(c)

list_ = soup.select("div h2")
for c in list_:
    print(c)
```

```python
def fetch_titles(url: str):
    # ネットワーク取得（失敗時の例外は上位でキャッチする想定）
    resp = requests.get(url, timeout=15)
    resp.raise_for_status()

    # Shift_JISなど日本語ページの文字コード対策
    # requestsは大抵自動判定するが、ずれている場合はencodingを明示
    if not resp.encoding or resp.encoding.lower() == 'iso-8859-1':
        resp.encoding = resp.apparent_encoding

    soup = BeautifulSoup(resp.text, "html.parser")

    # 指定階層: <html> → <body> → <center> → ... → <strong>
    # center直下やその下にあるstrongを対象にする
    strong_tags = soup.select("html > body > center strong")

    # タイトル抽出
    titles = []
    for tag in strong_tags:
        text = tag.get_text(strip=True)
        if text:
            titles.append(text)
    return titles

if __name__ == "__main__":
    url = "https://abehiroshi.la.coocan.jp/movie/eiga.htm"  # 取得したいページURLに置き換え
    titles = fetch_titles(url)
    for t in titles:
        print(t)
```

```python
#武蔵野大学のWebサイトのサイトマップを抽出する
    #武蔵野大学Webサイトのトップページにアクセス
    #同一ドメインの全てのリンク（コメントアウトされていないもの）を辿り，全ページのURLと<title>を辞書型変数に格納する
        #key：URL
        #value：<title></title>で挟まれた文字列
#辞書型変数を print() で表示する
# Webサイトにアクセスする際は必ず time.sleep(秒数) を使って負荷軽減をすること

import time
from urllib.parse import urljoin, urlparse

# 基本URL
base_url = "https://www.musashino-u.ac.jp/"

# サイトマップを格納する辞書
sitemap = {}

# 訪問済みURLを記録するセット
visited_urls = set()

# 訪問待ちのURLを格納するリスト
urls_to_visit = [base_url]

# ユーザーエージェントを設定
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# 同一ドメインかどうかを判定する関数
def is_same_domain(url, base_url):
    """URLが同一ドメインかどうかを判定"""
    base_domain = urlparse(base_url).netloc
    url_domain = urlparse(url).netloc
    return base_domain == url_domain

# クロール開始
print("クロールを開始します...")
print(f"基準URL: {base_url}\n")

while urls_to_visit:
    # 次のURLを取得
    current_url = urls_to_visit.pop(0)
    
    # すでに訪問済みならスキップ
    if current_url in visited_urls:
        continue
    
    # 訪問済みとしてマーク
    visited_urls.add(current_url)
    
    try:
        # HTTPリクエストを送信
        print(f"アクセス中: {current_url}")
        res = requests.get(current_url, headers=headers, timeout=10)
        res.encoding = res.apparent_encoding
        
        # ステータスコードが200でない場合はスキップ
        if res.status_code != 200:
            print(f"  → ステータスコード {res.status_code}: スキップ")
            continue
        
        # HTMLを解析
        soup = BeautifulSoup(res.text, 'html.parser')
        
        # <title>タグを取得
        title_tag = soup.find('title')
        title = title_tag.string.strip() if title_tag and title_tag.string else "タイトルなし"
        
        # 辞書に格納
        sitemap[current_url] = title
        print(f"  → タイトル: {title}")
        
        # ページ内の全ての<a>タグを取得
        a_tags = soup.find_all('a', href=True)
        
        for a in a_tags:
            href = a.get('href')
            
            # 絶対URLに変換
            absolute_url = urljoin(current_url, href)
            
            # フラグメント（#以降）を除去
            absolute_url = absolute_url.split('#')[0]
            
            # 同一ドメインで、まだ訪問していないURLを追加
            if (is_same_domain(absolute_url, base_url) and 
                absolute_url not in visited_urls and 
                absolute_url not in urls_to_visit):
                urls_to_visit.append(absolute_url)
        
    except Exception as e:
        print(f"  → エラー: {e}")
        continue

print(f"\n\nクロール完了!")
print(f"訪問したページ数: {len(sitemap)}")
print(f"\n{'='*80}")
print("サイトマップ（辞書型変数）:")
print(f"{'='*80}\n")

# 辞書を表示
for url, title in sitemap.items():
    print(f"URL: {url}")
    print(f"Title: {title}")
    print("-" * 80)
```
